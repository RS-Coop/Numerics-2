```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
```

```python
'''
a = [a_m ..., a_0]
b = same as above
where the terms match [Exp[i*m*theta], Exp[i*(m-1)*theta, ..., 1]

theta in [0,2pi]
'''
def ros(theta, a, b):
    s = len(a)
    
    assert s == len(b)
    
    #Convert to numpy
    a = np.array(a)
    b = np.array(b)
    
    #Compute A, C, B, D
    
    A = np.dot(a, np.array([np.cos((s-k-1)*theta) for k in range(s)]))
    B = np.dot(a, np.array([np.sin((s-k-1)*theta) for k in range(s)]))
    
    C = np.dot(b, np.array([np.cos((s-k-1)*theta) for k in range(s)]))
    D = np.dot(b, np.array([np.sin((s-k-1)*theta) for k in range(s)]))
    
    #Check these
    mu_r = (A*C+B*D)/(C**2+D**2)
    mu_i = (B*C-A*D)/(C**2+D**2)
    
    return mu_r, mu_i
```

```python
theta = np.linspace(0, 2*np.pi, 1000)
```

```python
ab2 = ([1, -1, 0], [0, 3/2, -1/2])
ab3 = ([1, -1, 0, 0], [0, 23/12, -4/3, 5/12])
am2 = ([1, -1, 0], [5/12, 2/3, -1/12])
bd2 = ([1, -4/3, 1/3], [2/3, 0, 0])
bd3 = ([1, -18/11, 9/11, -2/11], [6/11, 0, 0, 0])
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,10))

mu_r, mu_i = ros(theta, ab2[0], ab2[1])

lim = 1

ax.plot(mu_r, mu_i)
ax.hlines(0,-lim,lim,'k')
ax.vlines(0,-lim,lim,'k')

ax.set_title('Region of Absolute Stability')
ax.set_xlabel('Real Axis')
ax.set_ylabel('Imaginary Axis')

ax.set_xlim(-lim,lim)
ax.set_ylim(-lim,lim)

ax.grid(False)
```
